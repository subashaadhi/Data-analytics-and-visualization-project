```python
df.isnull().sum()

age_mean=df["age"].mean()
age_mean

df["age"].fillna(age_mean,inplace=True)

df
```

```python
df.isnull().sum()

# Calculate the mode of the 'job' column
job_mode = df['job'].mode()[0]

# Fill missing values in the 'job' column with the mode
df['job'].fillna(job_mode, inplace=True)

df.isnull().sum()
```

```python
df.isnull().sum()

df['duration'].value_counts()

max_counts=df['duration'].value_counts().index[0]
max_counts

df['duration'].fillna(max_counts,inplace=True)
```
